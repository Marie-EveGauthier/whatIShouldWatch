```python
#Get a list of ALL categories
cats = data.categories
all_cats = []
for cat in cats:
    split= str(cat).split(',')
    for sub_cat in split:
        replacables = ["\\'Category:", "Category:", '"', "u'", "'", '[[', ']]', '[', ']']
        for repl in replacables:
            sub_cat = sub_cat.replace(repl, "")
        all_cats.append(sub_cat)

print all_cats[:10]
```

```python
unique = list(set(all_cats))
print unique
```
